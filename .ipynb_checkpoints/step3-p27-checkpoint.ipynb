### HSV的色彩空间
HSV(Hue,Saturation,Value)
是根据颜色的直观特性由A.R.Smith在1978年创建的-种颜色空间,也称六角锥体模型(HexconeModel)
这个模型中颜色的参数分别是:色调(H),饱和度*(S)，亮度(V)。  
- 色调H:用角度度量，取值范围为0° ~ 360°， 从红色开始按逆时针方向计算,红色为0°，绿色为
120*,蓝色为240°。它们的补色是:黄色为60°，青色为180°,品红为300°
- 饱和度S:取值范围为0.0~ 1.0
- 亮度V:取值范围为0.0(黑色) ~ 1.0(白色)

RGB和CMY颜色模型都是面向硬件的，而HSV (Hue Saturation Value)颜色模型是面向用户的

### opencv中HSV取值范围
在opencv中颜色的参数及取值范围分别是:
- 色调Hue: [0,179]
- 饱和度Saturation: [0,255]
- 亮度Value: [0,255]

不同的软件使用不同的取值范围，为了对比，通常需要对这些范围进行标准化



```python
%matplotlib inline 
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tool

tool.setGrid(1,2)

frame = cv.imread('./img/Eiffel.jpeg',1) #彩色图像
cv.circle(frame, (100,100), 15, (255,0,0), -1)
imageRGB = cv.cvtColor(frame, cv.COLOR_BGR2RGB )
tool.pltGrid(1,1)
plt.imshow( imageRGB )

hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)# 转换像素编码格式
##可以使用cv.imshow在窗口显示
# cv.imshow( 'hsv', hsv)
# k=cv.waitKey(0) & 0xFF
# cv.destroyAllWindows()
#用matplotlib显示
# imageRGB = cv.cvtColor(hsv, cv.COLOR_BGR2RGB ) # 这样显示没意义
imageRGB = cv.cvtColor(hsv, cv.COLOR_HSV2RGB )
tool.pltGrid(1,2)
plt.imshow( imageRGB )
plt.show( )

#确认换色灯光的HSV值范周
lower_blue = np.array([0,0,81]) #hsv
upper_blue = np.array([115,255,255])
#使用闽值，确定范围
mask = cv.inRange(hsv, lower_blue, upper_blue)
# print(mask[100])## mask的元索值，是0或255
imageRGB = cv.cvtColor(mask, cv.COLOR_BGR2RGB )
tool.pltGrid(1,1)
plt.imshow( imageRGB)

res = cv.bitwise_and(frame, frame, mask=mask)
imageRGB = cv.cvtColor(res, cv.COLOR_BGR2RGB)
tool.pltGrid(1,2)
plt.imshow( imageRGB )
plt.show( )

```

## P29 30 滑块确定色彩取值范围的神器

ps 相关代码在上一P视频的末尾

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# HSV色彩空间图
frame = cv.imread('./img/hsv.png',1)
##铁塔图像
# frame = cv.imread('./img/Eiffel.jpeg',1)
# cv.circle(frame, (100, 100), 15, (255,0,0), -1)
###练习题图像
# frame = cv. imread('./img/color.webp',1) #七彩图像
# cv.imshow( "BGR",frame) #显示图片
#设置范围初始值
hsv_low = np.array([0, 0, 0])
hsv_high = np.array([120, 122, 255])

#下面几个两数，写得有点冗余
def h_low(value):
    hsv_low[0] = value
def h_high(value):
    hsv_high[0] = value
def s_low(value):
    hsv_low[1] = value
def s_high(value):
    hsv_high[1] = value
def v_low(value):
    hsv_low[2] = value 
def v_high(value):
    hsv_high[2] = value

cv.namedWindow( 'image' ,cv.WINDOW_NORMAL)
# cv.namedWindow( 'dst' ,cv.WINDOW_NORMAL)
cv.namedWindow( 'result' ,cv.WINDOW_NORMAL)
cv.createTrackbar('H low', 'image', 0,180, h_low)
cv.createTrackbar('H high','image', 0,180, h_high)
cv.createTrackbar('S low', 'image', 0,255, s_low)
cv.createTrackbar('S high','image', 0,255, s_high)
cv.createTrackbar('V low', 'image', 0,255, v_low)
cv.createTrackbar('V high','image', 0,255, v_high)

cv.setTrackbarPos('H low', 'image',hsv_low[0])
cv.setTrackbarPos('H high','image',hsv_high[0])
cv.setTrackbarPos('S low', 'image',hsv_low[1])
cv.setTrackbarPos('S high','image',hsv_high[1])
cv.setTrackbarPos('V low', 'image',hsv_low[2])
cv.setTrackbarPos('V high','image',hsv_high[2])

while True:
    dst = cv.cvtColor(frame, cv.COLOR_BGR2HSV) # BGR转HSV
    dst = cv.inRange(dst, hsv_low, hsv_high) #通过HSV的高低阀值，提取图像部分区龙
    # cv.imshow( 'dst', dst)
    result = cv.bitwise_and(np.array(frame),frame,mask=dst )
    cv.imshow( 'result', result)
    r = cv.getTrackbarPos('H low', 'image')
    r = cv.getTrackbarPos('H high','image')
    r = cv.getTrackbarPos('S low', 'image')
    r = cv.getTrackbarPos('S high','image')
    r = cv.getTrackbarPos('V low', 'image')
    r = cv.getTrackbarPos('V high','image')
    if cv.waitKey(1) & 0xFF == 27: # ord('q'):
        break
cv . destroyAllWindows( )
```

## P31 习题：同时筛选多个色彩

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tool
%matplotlib inline
tool.setGrid(1,2)

# frame = cv.imread('./img/hsv.png',1) #彩色图像
frame = cv.imread('./img/color.webp',1) #彩色图像
#转换到HSV
hsv=cv.cvtColor( frame , cv.COLOR_BGR2HSV)

lower_red_1=np.array([166, 100,100])
upper_red_1=np.array([180,255,255])
lower_red_2=np.array([0,100, 100])
upper_red_2=np.array([4,255,255])

lower_yellow_green=np.array([20, 100, 100])
upper_yellow_green=np.array([94, 255,255])

#根据阈值构建掩模
mask_red_1=cv.inRange(hsv, lower_red_1, upper_red_1)
mask_red_2=cv.inRange(hsv, lower_red_2, upper_red_2)
mask_yellow_green=cv.inRange(hsv, lower_yellow_green, upper_yellow_green)

# mask合并
mask_red = cv.bitwise_or(mask_red_1 ,mask_red_2)
imageRGB = cv.cvtColor(mask_red, cv.COLOR_BGR2RGB )
tool.pltGrid(1,1)
plt.imshow( imageRGB)
# plt.show( )

mask = cv.bitwise_or(mask_red, mask_yellow_green)
imageRGB = cv.cvtColor( mask, cv.COLOR_BGR2RGB )
tool.pltGrid(1,2)
plt.imshow( imageRGB)
plt.show( )

#对原图像和掩模进行位运算
res=cv.bitwise_and( frame, frame , mask=mask)
#显示图像
imageRGB = cv.cvtColor(res, cv.COLOR_BGR2RGB )
plt.imshow( imageRGB)
plt.show( )

```

## P32 图像变换缩放

OpenCV提供了两种图像转换的函数，cv.warpAffine和cv.warpPerspective

<br/>

- API  
dst=cv.resize(src, dsize\[, dst\[, fx\[, fy\[, interpoltio\]\]\]\])

- 计算公式  
$ dst(x, y) = src(M_{11}x + M_{12}y+ M_{13}, M_{21}x + M_{22}y + M_{23}) $

- 参数  
src: 输入图像  
dst: 输出图像，与输入图像形状相同  
dsize: 输出图像数据的大小，等于0的时候，计算方法为  
$ dsize = Size(round(f_x * sre.cols), round(f, * sre.rows) $  
$f_x$:扩大的倍数  
$ (double)dsize.width= f_x * src.cols $  
$f_y$:扩大的倍数  
$ (double)dsize. width= f_y * sre. cols $  



```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tool
%matplotlib inline

img = cv.imread('./img/PorteDeLaGrosseHorloge.jpg',1) #彩色图像
imageRGB = cv.cvtColor( img, cv.COLOR_BGR2RGB )
plt . imshow( imageRGB)
plt. show()

res = cv.resize(img, None, fx=2, fy=2, interpolation = cv.INTER_CUBIC)
imageRGB = cv.cvtColor(res, cv.COLOR_BGR2RGB)
plt.imshow( imageRGB )
plt.show()


height, width = img.shape[ :2]
res = cv.resize(img, (2*width, 2*height), interpolation = cv.INTER_CUBIC)
imageRGB = cv.cvtColor(res, cv.COLOR_BGR2RGB)
plt.imshow( imageRGB )
plt.show()

```

## P33 平移

- 通式  
dst=cv.warpAffine(src, M, dsize[, dst[, flags, borderMode[, bordervalel1])
- 计算公式  
$ dst(x, y)= sre(M_{11}x + M_{12}y + M_{13}. M_{21}x + M_{22}y + M_{23}) $
- 参数  
src: 输入图像  
dst: 输出图像  
M: 2x3 矩阵  
$ M=
\begin{pmatrix}
M_11 & M_12 & M_13 \\
M_21 & M_22 & M_23
\end{pmatrix}
$  
dsize:输出图像数据的大小，注意这里要转换一下  
flags:计算差值的方法  

移动矩阵:  
$ M=
\begin{pmatrix}
1 & 0 & t_x \\
0 & 1 & t_y
\end{pmatrix}
$



```python
import numpy as np
import cv2 as cv
img = cv.imread('./img/PorteDeLaGrosseHorloge.jpg',0) #黑白图像
rows, cols = img.shape ##注意这里
M = np.float32([[1,0,60],[0,1,80]])
dst = cv.warpAffine(img,M, (cols,rows)) #注意第三个参数，顺序
imageRGB = cv.cvtColor(dst, cv.COLOR_BGR2RGB)
plt.imshow( imageRGB)
plt.show( )
```

## P34 图像旋转效果

getRotationMatrix2D()

- API  
retval=cv.getRotationMatrix2D(center, angle, scale)
- 计算公式  
$
\begin{bmatrix}
α & β & (1-α)\cdot center.x - β\cdot center.y \\
-β & α & β\cdot center.x+(1- α)\cdot center.y
\end{bmatrix}
$  
其中:  
$
α = scale \cdot \cos angle, \\
β= scale \cdot \sin angle
$
  
- 参数  
center:旋转中心  
angle:旋转角度，正数代表逆时针旋转，左上角坐标默认(0, 0)  
scale:各个方向同等扩大比例  


```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tool
%matplotlib inline

img = cv.imread('./img/PorteDeLaGrosseHorloge.jpg',0) #黑白图像
rows, cols = img.shape
# cols-1 and rows-1 are the coordinate 1 imits.
M = cv.getRotationMatrix2D(((cols-1)/2.0, (rows-1)/2.0),60,1) # 旋转中心 角度 放大
dst = cv.warpAffine( img,M, (cols, rows))

print(M,dst.shape)

imageRGB = cv.cvtColor(dst, cv.COLOR_GRAY2BGR)#cv.COLOR_BGR2RGB )
plt.imshow( imageRGB)
plt.show()
```

## P35 图像旋转原理


```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tool
%matplotlib inline

tool.setGrid(1,3)

#生成一张黑色的图像
img=np.zeros((1000, 1000,3),np.uint8)
rows,cols,_ = img.shape
#画出线段观察坐标系
cv.line(img,(0,0),(200, 500),(255,200,0),10) #蓝色线段
cv.circle(img,(200, 500),20, (255, 100,255),-1)
#显示图像
imageRGB = cv.cvtColor( img, cv.COLOR_BGR2RGB )
tool.pltGrid(1,1)
plt.imshow( imageRGB)

M = cv.getRotationMatrix2D((0, 0),30,1)
dst = cv.warpAffine(img, M, (cols, rows))
imageRGB = cv.cvtColor(dst, cv.COLOR_BGR2RGB )
tool.pltGrid(1,2)
plt.imshow( imageRGB)

#画出线段观察坐标系
cv.line(dst,(0,0), (200, 500), (255,200,0),10)
cv.circle(dst, (200, 500),20, (255, 100,255),-1)
cv.ellipse(dst,(0,0),(150, 150),0,37,65, (255, 100,255), 10)
cv.ellipse(dst, (0,0),(250,250),0,0,65, (100,250,255),10)

font=cv.FONT_HERSHEY_SIMPLEX
cv.putText(dst, 'A(x,y)',(200, 600),font, 2, (255,255,255),7)
cv.putText(dst,"A'(x' ,y')",(500,450),font, 2, (255,255,255),7)
cv.putText(dst,"a", (110,180),font, 3,(255, 100,255),7)
cv.putText(dst,"b", (260, 100),font, 3,(100,250,255),7)
cv.putText(dst,"r",(80,400),font, 3, (255,200,0),5)
cv.putText(dst,"r", (360,220),font, 3,(255,200,0),5)
#显示图像
tool.pltGrid(1,3)
imageRGB = cv.cvtColor(dst, cv.COLOR_BGR2RGB)
plt.imshow( imageRGB)
plt.show()
```

整理，写成矩阵的形式  
$
M =
\begin{pmatrix}
n\cos α & n\sin α & x_0(1 - n\cos α) - y_0(n\sin α)\\
- n\sinα & n\cos α & y_0(1 - n\cos α)+ x_0(n\sinα)
\end{pmatrix}
$

cv.getRotationMatrix2D()，就是得到这个矩阵  
作为参数传入cv.warpAffine(img,M,(cols,rows))，与原图像点坐标进行计算



## P36 仿射变换透规变换
在仿射变换中，原图像中的所有平行线在输出图像中仍然是平行的  
为了找到变换矩阵，我们需要从输入图像中选取三个点，以及它们在输出图像中的对应位置


```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tool
%matplotlib inline
# tool.setGrid(1,1)

#仿射变换

img = cv.imread('./img/PorteDeLaGrosseHorloge.jpg',1) #黑白图像
rows,cols,ch = img.shape
# cv.circle( img, (400, 200),20, (255, 100,255),-1)
# cv.circle(img, (400, 600),20, (255, 100,255),-1)
# cv.circle(img, (800,200),20, (255, 100,255),-1 )
cv.circle( img, (200, 100),20, (255, 100,255),-1)
cv.circle(img, (200, 300),20, (255, 100,255),-1)
cv.circle(img, (400,100),20, (255, 100,255),-1 )
imageRGB = cv.cvtColor( img, cv.COLOR_BGR2RGB )
plt.imshow( imageRGB)
plt.show()

# pts1 = np.float32([[400, 200], [400, 600], [800, 200]])
# pts2 = np.float32([[200, 400], [600, 600], [800,250]])
pts1 = np.float32([[200, 100], [200, 300], [400, 100]])
pts2 = np.float32([[100, 200], [300, 300], [400,125]])
M = cv.getAffineTransform(pts1,pts2) #得到变换矩阵
dst = cv.warpAffine( img,M, (cols,rows)) #进行变换
imageRGB = cv.cvtColor(dst, cv.COLOR_BGR2RGB )
plt.imshow( imageRGB)
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tool
%matplotlib inline
# tool.setGrid(1,1)

# 透规变换
img = cv.imread('./img/PorteDeLaGrosseHorloge.jpg',1) #彩色图像
rows, cols,ch = img.shape
imageRGB = cv.cvtColor ( img, cv.COLOR_BGR2RGB )
plt.imshow( imageRGB)
plt.show()
pts = np.array([[410,80], [900, 80], [900, 400], [410, 400]], np.int32 )
pts = np.array(pts/2, np.int32 )
print(pts)
cv.polylines(img,[pts], True, (0,255,255),10)
imageRGB = cv.cvtColor ( img,cv.COLOR_BGR2RGB )
plt.imshow( imageRGB)
plt.show()
pts1 = np.float32([[410, 80],[900, 80], [900, 400], [410, 400]])
# pts2 = np.float32([[10,10],[1200, 10],[1200, 900],[10, 900]])
pts2 = np.float32([[200,200],[1200, 30],[1600, 1200],[300, 900]])

pts1 = pts1/2
pts2 = pts2/2

M = cv.getPerspectiveTransform(pts1,pts2)
dst = cv.warpPerspective(img, M, (1300, 1000))
imageRGB = cv.cvtColor(dst, cv.COLOR_BGR2RGB)
plt.imshow( imageRGB)
plt.show()

```
