## P60 图像金字塔pyrDown效果

图像金字塔的概念  
使用图像金子塔创造一个新的水果"苹橙果子” 
主要函数: cv.pyrUp(), cv.pyrDown()  

保证图像融合的过渡 且不损失清晰度


通常情况下，使用的是固定大小的图像  
但是在某些情况下，我们需要在不同的分辨率下处理(相同的)图像  
例如，当在图像中搜索某个东西时，比如人脸，我们不确定该对象在该图像中的大小  
在这种情况下，我们需要创建一-组具有不同分辨率的相同图像，并在所有这些图像中搜索对象  
这些<u>具有不同分辨率的图像集</u>被称为图**像金字塔**  

有两种金字塔，高斯金字塔和拉普拉斯金字塔

- 高斯金字塔

高斯金字塔中的高阶(低分辨率)是通过去除低阶(高分辨率)图像中连续的行和列而形成的
我们可以使用cv.pyrDown()和cv.pyrUp()函数找到高斯金字塔。

## P62 pyrUp效果及原理

图像放大 

向上采样时，每个像素的右侧、下方分别插入零列和零行  

朐下采样时所使用的高斯滤波器对图像进行滤波  
由于面积扩大四倍，且增加的部分都是0元素,整体均值会降为原来的1/4  
故为了确保像素值区间在向上采样后与原始图像保持一致， 需要将高斯滤波器的系数乘以4

## P67 图像轮廓

找到图像轮廓，绘制图像轮廓  
主要函数: cv.findContours(), cv.drawContours()  

边缘检测虽然能够检测出边缘，但边缘是不连续的  
图像轮廓是指将边缘连接起来形成一个整体  
通过对图像轮廓的操作，我们能够获取目标图像的大小、位置、方向等信息  

查找轮廓cv.findContours()

contours, hierarchy=cv.findContours(image, mode, method\[,contours\[, hierarcy\[,offsetf\]\]\])
A
- contours: 返回轮廓
- hierarchay: 图像的拓扑信息(轮廓层析)
- image: 原始图像，8位单通道图像，所有非零值被处理为1,所有零值保持不变。也就是灰度图像会被自动处理为二值图像
- mode:轮廓检索模式
- method: 轮廓的近似方法

要求是8位单通道二值图像，
cv.cvtColor(imgray,cv.COLOR_BGR2RGB)
ret，thresh = cv.threshold(imgray,180, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE，cv.CHAIN_APPR0X_SIMPLE)

返回值: contours  
其中contours是个列表，其中包含-组轮廓信息， 每个轮廓由若干个点构成
